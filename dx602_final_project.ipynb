# SECTION 4: Descriptive Statistics

For this section, I think its important to get a high level view of the dataset. I'm not much of a hockey fan, so I want to get a better sense of who the best teams are, what their stats look like, and see what their coaching characteristics are.

The raw data doesnt really give a good sense of descriptive statistics since all of the stats are separated by home or away, we will worry about this later.

However, if you find the average difference between a given home stat vs away stat (seen below), you can see that generally the home team is able to perform better than the away team. In sports, its common for the home team to have a slight advantage, and it looks like hockey is no different.

```python
stats = ['goals', 'shots', 'hits', 'pim', 'powerPlayOpportunities', 
         'powerPlayGoals', 'giveaways', 'takeaways', 'blocked']

game_stats = pd.DataFrame()

for stat in stats:
    game_stats[f'diff_{stat}'] = df[f'home_{stat}'] - df[f'away_{stat}']


print("Home Team Advantage Table")
game_stats.describe().round(3)
```

Next, we need to aggregate the team stats so that they arent segregated by home and away. The separation doesn't mean they are useless in general, but because I want to get a high level look at the league, I am going to aggregate for simplicity. This view is going to give us a really insightful view on who the best teams are in the almost 20 years of data we have. For this view, we are going to specifically look at wins, loses and coach counts.

We want to look at the coach counts because the purpose of this project to see how the coach can affect a team's performance. In professional sports, if a team is not performing well, generally the first thing the organization does is fire their head coach to make a change. Therefore, I am going to make the assumption that a successful team is one that wins many games and has low coach turnover. 

```python
# Home Stats
home_wins = df.groupby('home_team').agg(
    home_won=('home_won', 'sum'),
    away_won=('away_won', 'sum')
).reset_index().rename(columns={'home_team': 'team'})

# Away Stats
away_wins = df.groupby('away_team').agg(
    home_won=('home_won', 'sum'),
    away_won=('away_won', 'sum')
).reset_index().rename(columns={'away_team': 'team'})

# Total Stats
team_wins = pd.concat([home_wins, away_wins]).groupby('team').sum().reset_index()

# Same logic for coaches
home_coaches = df[['home_team', 'home_head_coach']].rename(
    columns={'home_team': 'team', 'home_head_coach': 'coach'})
away_coaches = df[['away_team', 'away_head_coach']].rename(
    columns={'away_team': 'team', 'away_head_coach': 'coach'})

#total Coach Counts
coach_counts = (pd.concat([home_coaches, away_coaches]).groupby('team')['coach'].nunique().reset_index().rename(columns={'coach': 'home_head_coach'}))

#Join them together
team_wins = team_wins.merge(coach_counts, on='team', how='left')

# Calculate statistics
team_wins['total_wins'] = team_wins['home_won'] + team_wins['away_won']
team_wins['avg_wins_per_coach'] = (team_wins['total_wins'] / team_wins['home_head_coach']).round(2)
team_wins.sort_values(by='total_wins', ascending=False)
```

Now that I have a better view of the teams, I can find the average number of coaches per team, as well as the average win count over the course of the league.

After calculating this, I can find the 'successful' teams by seeing which ones have less coach turnover and more total wins.

```python
mean_coaches = team_wins['home_head_coach'].mean().round()
avg_total_wins = team_wins['total_wins'].mean().round()

print(f"Average number of coaches per team: {mean_coaches}")
print(f"Average total wins: {avg_total_wins}")
print()

best_teams = team_wins.query('`total_wins` >= @avg_total_wins and `home_head_coach` <= @mean_coaches')
print(f"SUCCESSFUL NHL TEAMS: teams with > {avg_total_wins} total wins and < {mean_coaches} head coaches ")
best_teams
```

After filtering the data, I am able to identify 13 teams that would be considered succesful.

From this list, I can make another assumption that if you have ever coached one of these teams, you would be considered a well-performing coach.

For the last part of this section I will display the list of successful coaches

```python
best_teams = best_teams['team'].tolist()
home_coaches = df[df['home_team'].isin(best_teams)][['home_team', 'home_head_coach']].rename(
    columns={'home_team': 'team', 'home_head_coach': 'coach'})

# Get coaches from away games
away_coaches = df[df['away_team'].isin(best_teams)][['away_team', 'away_head_coach']].rename(
    columns={'away_team': 'team', 'away_head_coach': 'coach'})

best_coaches = pd.concat([home_coaches,away_coaches]).drop_duplicates()
best_coaches['coach'].unique()
```

```python
# Home Stats by coach
home_wins = df.groupby('home_head_coach').agg(
    home_won=('home_won', 'sum'),
    away_won=('away_won', 'sum')
).reset_index().rename(columns={'home_head_coach': 'coach'})

# Away Stats by coach
away_wins = df.groupby('away_head_coach').agg(
    home_won=('home_won', 'sum'),
    away_won=('away_won', 'sum')
).reset_index().rename(columns={'away_head_coach': 'coach'})

# Total Stats by coach
coach_wins = pd.concat([home_wins, away_wins]).groupby('coach').sum().reset_index()

# Count distinct teams per coach
home_teams = df[['home_head_coach', 'home_team']].rename(
    columns={'home_head_coach': 'coach', 'home_team': 'team'})
away_teams = df[['away_head_coach', 'away_team']].rename(
    columns={'away_head_coach': 'coach', 'away_team': 'team'})

# Total team counts per coach
team_counts = (pd.concat([home_teams, away_teams]).groupby('coach')['team'].nunique().reset_index().rename(columns={'team': 'teams_coached'}))

# Join together
coach_wins = coach_wins.merge(team_counts, on='coach', how='left')

# Calculate statistics
coach_wins['total_wins'] = coach_wins['home_won'] + coach_wins['away_won']
coach_wins['avg_wins_per_team'] = (coach_wins['total_wins'] / coach_wins['teams_coached']).round(2)
coach_wins.sort_values(by='total_wins', ascending=False)

coach_wins.sort_values('total_wins', ascending=False)
```

Finally, I am going to create the dataframe that will be used for model creation. This is going to be the same aggregation of stats as the previous team_stats table, but this one will be grouped by year. 

Using this, we can visualize team performance over time and get better insights to a team's journey over the past 18 years of games.

```python
def create_yearly_team_df(df):
    stats = ['goals', 'shots', 'hits', 'pim', 'powerPlayOpportunities',
             'powerPlayGoals', 'giveaways', 'takeaways', 'blocked']

    # Build agg dict for home and away perspectives
    home_agg = {'home_won': 'sum', 'away_won': 'sum'}
    away_agg = {'home_won': 'sum', 'away_won': 'sum'}

    for stat in stats:
        home_agg[f'home_{stat}'] = 'sum'
        home_agg[f'away_{stat}'] = 'sum'
        away_agg[f'home_{stat}'] = 'sum'
        away_agg[f'away_{stat}'] = 'sum'

    # Home stats (team was home)
    home_df = df.groupby(['home_team', 'year', 'home_head_coach']).agg(home_agg).reset_index()
    home_df.rename(columns={'home_team': 'team', 'home_head_coach': 'coach'}, inplace=True)

    # Away stats (team was away)
    away_df = df.groupby(['away_team', 'year', 'away_head_coach']).agg(away_agg).reset_index()
    away_df.rename(columns={'away_team': 'team', 'away_head_coach': 'coach'}, inplace=True)

    # Combine and sum across home/away appearances
    combined = (pd.concat([home_df, away_df]).groupby(['team', 'year', 'coach']).sum().reset_index())

    # Derive totals
    combined['total_wins'] = combined['home_won'] + combined['away_won']
    for stat in stats:
        combined[f'total_{stat}'] = combined[f'home_{stat}'] + combined[f'away_{stat}']

    # Drop intermediate home/away columns if not needed
    cols_to_drop = [f'home_{s}' for s in stats] + [f'away_{s}' for s in stats] + ['home_won', 'away_won']
    combined.drop(columns=cols_to_drop, inplace=True)

    return combined

yearly_stats = create_yearly_team_df(df)
yearly_stats.sort_values('total_wins', ascending=False)
```

# SECTION 12: Linear Regression

In this section we will finally test to see if the coach can truly make a difference in the amount of wins a team will get in a season.

We will start by defining the feature and target variables, removing the non numeric columns.

```python
import sklearn.linear_model
import numpy as np

# filter out non numeric columns
features = np.array(yearly_stats['coach_encoded']).reshape(-1,1)
target = np.array(yearly_stats['total_wins'])

def regression_model(features, target):
    ''' builds and scores a linear regression model with the features and targets provided '''
    model = sklearn.linear_model.LinearRegression()
    model.fit(features,target)
    print("SCORE",model.score(features,target).round(2))

    predictions = model.predict(features)

    plt.scatter(target, predictions, label='Predictions')
    plt.plot([target.min(), target.max()], [target.min(), target.max()], 
         color='red', linewidth=2, label='Regression Line')
    plt.xlabel('Actuals')
    plt.ylabel('Predictions')
    plt.title('Actuals vs Predictions')
    plt.legend()



    return predictions 

preds = regression_model(features,target)

```

The helper function creates the linear regression model and fits it to the data provided.

With a score of .24, it can safely be said that, in the data's current form, there is no relationship between the coach and the team's wins for a season.

For fun, we will create another model, using multivariable regression, using all of the stats in the dataset to see if they can predict the total wins better

```python
features_all = yearly_stats.drop(columns=['total_wins','team','coach'])

second_model = regression_model(features_all, target)
```
